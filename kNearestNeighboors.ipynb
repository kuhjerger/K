```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
x = np.linspace(0,10, N)
y = x**2 + 3*np.random.randn(N)+20
plt.scatter(x,y)
```

```python
x = x.reshape(N,1)
y = y.reshape(N,1)
```

```python
xnew = np.linspace(0,10, 200)
xnew = xnew.reshape(len(xnew),1)
```

```python
def d_squared(u,v):
    l =[]
    for i in range(0,len(v)):
        l.append((u-v[i]).T.dot(u-v[i]))
    l = np.asarray(l)
    l = l.reshape(len(l), 1)
    return l
```

```python
def nearestNeighboor(xnew,x,y):
    predictedY = []
    for i in range(0, len(xnew)):
        predictedY.append(y[np.argmin(d_squared(xnew[i], x))])
    return predictedY
```

```python
predictedY = nearestNeighboor(xnew,x,y)
```

```python
plt.scatter(x,y)
plt.plot(xnew,predictedY, color = "red")
```

```python
def kNearestNeighboors(xnew,x,y,K):
    predictedY = []
    for i in range(0, len(xnew)):
        temp = np.argsort(d_squared(xnew[i], x), axis = 0)[:K]
        predictedY.append(np.mean(y[temp]))
        
    return predictedY
```

```python
K = 30
predictedY = kNearestNeighboors(xnew,x,y,K)
```

```python
plt.scatter(x,y)
plt.plot(xnew,predictedY, color = "red")
```
